```python
import requests

w = 1

params={"view": "mRoster"}

url = 'http://fantasy.espn.com/apis/v3/games/ffl/seasons/2022/' \
'segments/0/leagues/{}'.format(leagueID,w,w)



response = requests.get(url, cookies=cookies, params = params)
```

```python
data = json.loads(response.text)
```

```python
pd.json_normalize(data.get('teams')[0].get('roster').get('entries')[0])
```

```python
rosters = pd.DataFrame()
for team in data.get('teams'):
    tid = team.get('id')
    roster = team.get('roster').get('entries')
    for player in roster:
        df = pd.json_normalize(player)
        df['teamID'] = tid
        rosters = rosters.append(df)
```

```python
espn = rosters[['teamID','playerId']]
```
